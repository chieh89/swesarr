# Changes in autumn soil and snow conditions

- Freezing from end of October to November 20 with snow depths ranging between ~5-15 cm
- snow temperatures even out with snow accumulation >= 15 cm (e.g., end of October)
- ~30 cm snow accumulation from Nov 19–22 modulates a month of unfrozen soil
- snow continues accumulating in 3 big events on 22nd, 26th, 30th
- after Nov 20, soil temps hover within 1˚C of freezing for both 5 cm and 20 cm depth

## Implications for radar

- Radar flights on Nov 4–6, 2019
- variable soil temps with some surface-level freezing and minimal surface soil moisture
- compared to the February 10-11, 2020 flights
    - temperatures are similar,
    - soil moisture at depth (20 cm) is similar to the autumn (~0.10),
    - but surface soil moisture (measured in water fraction by volume) is higher by ~0.01 than in the autumn
