```python
from pathlib import Path
import pandas as pd
import geopandas as gpd
import pyproj
import datetime

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import hvplot.pandas
import holoviews as hv
```

```python
filepath = Path('~/shared-public/swesarr_project/nsidc-pulls/manual_depths/SNEX20_A19_SD_GM_201911_v01.csv')
radar_roi = Path('~/shared-public/swesarr_project/base-layers/swesarr-roi.shp')
```

```python
# open A19 snow depth data
df = pd.read_csv(filepath, header=2)

# convert UTM to Lat/Lon
transformer = pyproj.Transformer.from_crs("EPSG:32612", "EPSG:4326") # UTM Zone 12N 
df['latitude'], df['longitude'] = zip(*df.apply(lambda row: transformer.transform(row['Easting (m)'], row['Northing (m)']), axis=1))
df.head()
```

```python
radar_bbox = gpd.read_file(radar_roi)
```

```python
df.hvplot.scatter(x='longitude', y='latitude', by='#Date (YYYYMMDD)')
```
